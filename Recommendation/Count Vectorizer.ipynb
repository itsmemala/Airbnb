```python
import pandas as pd
data1=pd.read_csv('D:\\PGP-BABI\\Capstone\\Airbnb\\Tasmania-detailed listings - final2.csv', error_bad_lines=False)
data2=pd.read_csv('D:\\PGP-BABI\\Capstone\\Airbnb\\Sentiment Scores\\All_listing_feature_sentiment - Aggr3.csv', error_bad_lines=False)
data=pd.merge(data1,data2,on='listing_id',how='inner')
```

```python
data.head()
```

```python
col_names = data.columns.values.tolist()
data.loc[:, 'merged'] = ""
for each_col_ind in range(1,30):
    print('Concatenating', col_names[each_col_ind])
    data.loc[:, 'merged'] = data.loc[:, 'merged'] + " " + data[col_names[each_col_ind]].astype(str)
```

```python
data['merged'].head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# instantiating and generating the count matrix
count = CountVectorizer()
count_matrix = count.fit_transform(data['merged'])

# generating the cosine similarity matrix
cosine_sim = cosine_similarity(count_matrix, count_matrix)
```

```python
# creating a Series for the movie titles so they are associated to an ordered numerical
# list I will use in the function to match the indexes
indices = pd.Series(data.listing_id)

#  defining the function that takes in listing id 
# as input and returns the top 10 recommended listings
def recommendations(listing, cosine_sim = cosine_sim):
    
    # initializing the empty list of recommended listings
    recommended_listings = []
    
    # gettin the index of the listing that matches the listing id
    idx = indices[indices == listing].index[0]

    # creating a Series with the similarity scores in descending order
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)

    # getting the indexes of the 10 most similar listings
    top_10_indexes = list(score_series.iloc[1:11].index)
    
    # populating the list with the best 10 matching listings
    for i in top_10_indexes:
        recommended_listings.append(list(data.listing_id)[i])
        
    return recommended_listings
```
