```python
import pandas as pd
data=pd.read_csv('D:\\GitHub\\Airbnb\\Feature Mapping + Sentiment\\All_listing_feature_sentiment_reviewer-Aggr1.csv', error_bad_lines=False)
```

```python
data.iloc[:,2:4]
```

```python
#This computes & stores feature wise sentiment score for each listing-reviewer combination
#final_data = pd.DataFrame([])
final_data = data.iloc[:,2:4]
loop = ["f0","f1","f2","f3","f4","f5","f6","f7"]
for i in range(len(loop)):
    non_zero = data.loc[data[loop[i]]!=0]
    #data_grouped = data.groupby('listing_id')[loop[i]].apply(lambda x: pd.Series([(x < 0).sum()/x.count(),(x > 0).sum()/x.count()])).unstack()
    data_grouped = non_zero.groupby(['listing_id', 'reviewer_id'])[loop[i]].apply(lambda x: pd.Series([x.mean()])).unstack()
    print(i)
    final_data = pd.merge(final_data,data_grouped[0], how = 'left', on = ['listing_id', 'reviewer_id'])
    #final_data[loop[i]+'p'] = data_grouped[1]
```

```python
final_data=final_data.drop_duplicates()
```

```python
final_data.to_csv('D:\\GitHub\\Airbnb\\Feature Mapping + Sentiment\\temp_featurewise.csv', sep=',')
```

```python
#This computes & stores overall sentiment score for each listing-reviewer combination
final_data = pd.DataFrame([])
non_zero = data.loc[data['compound_score_vader']!=0]
data_grouped = non_zero.groupby(['listing_id', 'reviewer_id'])['compound_score_vader'].apply(lambda x: pd.Series([x.mean()])).unstack()
final_data['overall_avg'] = data_grouped[0]
final_data.to_csv('D:\\GitHub\\Airbnb\\Feature Mapping + Sentiment\\temp_overall.csv', sep=',')
```
